#**Estimar por el metodo de Monte Carlo:**
$G=\int_{0}^{1} \cos{\frac{\pi x}{2}} dx $

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#Integral
a=0
b=1

#Definir la función
def f(x):
  return np.cos(np.pi*x/2)

#Numero de iteraciones
N=1000
```

```python
#Generar Npuntos aleatorios
x= np.random.uniform(a,b,N)
y= np.random.uniform(0,1,N)

#Puntos favorables
favorable= y<=f(x)
```

```python
#Grafica puntos
plt.figure(figsize=(10,6))
plt.plot(x[favorable], y[favorable], 'go', markersize=3, label='Puntos favorables')
plt.plot(x[~favorable], y[~favorable], 'ro', markersize=3, label='Puntos no favorables')

#Grafica funcion
x_funcion=np.linspace(a,b,N)
plt.plot(x_funcion, f(x_funcion), 'b-', label='f(x)=cos(πx/2)')
plt.title('Montecarlo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
