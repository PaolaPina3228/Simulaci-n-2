```python
import random
import matplotlib.pyplot as plt
```

```python
# Mapa de serpientes y escaleras
serpientes_escaleras = {13: 4, 17: 10, 3: 11, 15: 19}

def simular_juego():
  posicion = 1
  tiradas = 0
  posiciones=[posicion] #Guardamos posicion para grafica

  while posicion < 20:
    lanzar_dado = random.randint(1, 6)  # Tirar el dado
    posicion += lanzar_dado
    if posicion > 20:  # Si se pasa de 20, no avanza
        posicion = 20
    #Verificar si cae en serpientes o escaleras
    if posicion in serpientes_escaleras:
        posicion = serpientes_escaleras[posicion]

    posiciones.append(posicion)
    tiradas += 1

  return tiradas, posiciones
```

```python
#Simular un juego (Para probar el codigo)
t, p = simular_juego()
print(f"Tiradas: {t}")
print(f"Posiciones: {p}")

#Graficar el comportamiento de la ficha en un juego
plt.title("Comportamiento de la ficha")
x = range(t+1)
plt.plot(x, p, marker='o')
plt.xlabel('Tiradas')
plt.ylabel('Posicion')
plt.grid(True)
plt.show()
```

```python
# Simular múltiples juegos
def promedio_tiradas(N):
  total_tiradas = sum(simular_juego()[0] for _ in range(N))
  return total_tiradas / N

# Ejecutar simulación con 100,000 juegos
N = 100000
promedio = promedio_tiradas(N)
print(f"Promedio de tiradas: {promedio}")
```
